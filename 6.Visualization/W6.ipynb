```python
import numpy as np

def sigmoid(z):
    return 1./(1 + np.exp(-z))

z = np.linspace(-10, 10, 1000)
s = sigmoid(z)

plt.plot(z, s)
plt.title(r'$\sigma = \frac{1}{1+e^{-z}}$', fontsize = 24)
plt.xlabel('z', fontsize=20)
plt.ylabel(r'$\sigma$', fontsize=20)
plt.show()
```
