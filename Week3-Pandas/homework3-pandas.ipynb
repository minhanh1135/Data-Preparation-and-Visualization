```python
#1 magType 'ml' and type 'explosion'
len(dt[(dt['magType'] == 'ml') & (dt['type'] == 'explosion')])
```

```python
dt['distance'] = dt['place'].str.split('km').str[0]
dt.head(n=4)
```

```python
dt.distance.unique()
```

```python
c = dt[dt.distance.str.isnumeric()]
c[c['distance'] == '167']
```

```python
#2 Calculate the mean of mag in Alaska and the distance is smaller than 100km, example:
#"62km W of Cantwell, Alaska".
c[(c.place.str.contains('Alaska')) & ((c.distance.astype(int)) <= 100)]
```

```python
c[(c.place.str.contains('Alaska')) & ((c.distance.astype(int)) <= 100)]['mag'].mean()
```

```python
#3 Calculate the mean of mag in each quarter (q1:m1-m3, q2: m4-m6, q3: m7-m9, q4: m10-12), 
#the "time" field stores the milliseconds from 1/1/1970, hint: timestamp.
dt.head(n=2)
```

```python
pd.Timestamp(1539475168010, unit = 'ms')
```

```python
dt['Datetime'] = [pd.Timestamp(x, unit='ms') for x in dt.time]
dt.head(n=2)
```

```python
min_time = dt.Datetime.min()
max_time = dt.Datetime.max()
print('min', min_time, 'max_time', max_time)
```

```python
q = dt['Datetime'].quantile([0,0.25,0.5,0.75,1]).tolist()
q
```

```python
dt['q'] = pd.cut(dt['Datetime'], bins=q, labels=['q1', 'q2', 'q3', 'q4'])
dt.head(n=10)
```

```python
dt.groupby(dt['q'])['mag'].mean()
```
