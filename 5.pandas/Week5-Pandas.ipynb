```python
dat = pd.read_csv('/content/drive/MyDrive/ML/Data Visualization/L4/data/weather_by_station.csv')
dat.head()
```

```python
dat1 = dat[dat['datatype'] == 'SNOW'].groupby('station').mean()
dat1
```

```python
dat['date'] = pd.to_datetime(dat['date'], unit = 's')
```

```python
dat.info()
```

```python
dat['month'] = pd.DatetimeIndex(dat['date']).month
dat.head()
```

```python
pd.crosstab(
    index = dat.station_name,
    columns = dat.month,
    colnames= ['month'],
    values = dat.value,
    aggfunc= lambda x : (x >0).sum(),
)
```
