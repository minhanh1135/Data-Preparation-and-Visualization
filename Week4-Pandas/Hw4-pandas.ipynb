```python
covid = pd.read_csv('/content/drive/MyDrive/ML/Data Visualization/L4/data/covid19_cases.csv')
covid.head()
```

```python
#b, Create a date column using the data in the dateRep column and the pd.to_datetime() function.
covid['date'] = pd.to_datetime(covid['dateRep'])
covid.head()
```

```python
#c, set the date column as the index and sort the index
covid1 = covid.set_index('date')
covid1.sort_index()
```

```python
covid['countriesAndTerritories'].unique()
```

```python
#d, replace 
covid1['countriesAndTerritories'] = covid1['countriesAndTerritories'].replace('United_States_of_America','USA')
covid1['countriesAndTerritories'] = covid1['countriesAndTerritories'].replace('United_Kingdom','UK')
covid['countriesAndTerritories'].unique()
```

```python
#e 
covid1['countriesAndTerritories'].unique()
```

```python
c = ['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy',
     'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey','UK', 'USA']
```

```python
covid2 = covid[covid['countriesAndTerritories'].isin(c)]
covid2.head()
```

```python
covid2['countriesAndTerritories'].unique()
```

```python
covid2 = covid2.set_index('date')
covid2.sort_index()
covid2.tail()
```

```python
#f 
covid4 = covid[covid['countriesAndTerritories'].isin(c)]
pivot_covid4 = covid4.pivot(index='date', columns='countriesAndTerritories',
                            values= 'cases')
pivot_covid4.head()
```

```python
pivot_covid4.fillna(0)
```
